```python
# Download the list of S&P 500 tickers from Wikipedia
sp500_url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
sp500_table = pd.read_html(sp500_url)[0]
sp500_tickers = sp500_table['Symbol'].tolist()

# Define the start and end dates for the data
start_date = '2016-01-01'
#end_date = '2022-12-31'

#download prices data
data = yf.download(sp500_tickers, start=start_date)

# Keep only the 'Close' prices for each stock
prices = data['Close']

# Calculate daily returns for each stock
returns = prices.pct_change()
```

```python
data_spy = yf.download("SPY", start=start_date).reset_index()
data_GLD = yf.download("GLD", start=start_date).reset_index()
data_crude_oil =  yf.download('CL=F',start=start_date).reset_index()
data_vix =yf.download('^VIX',start=start_date).reset_index()
bitcoin_data = yf.download('BTC-USD', start=start_date).reset_index()
data_crude_oil['Close'].fillna(data_crude_oil['Close'].mean(), inplace=True)
```

```python
def calculate_MAPE(y, y_pred):
    y_true = np.array(y)
    y_pred = np.array(y_pred)
    mask = y_true != 0
    y_true_filtered = y_true[mask]
    y_pred_filtered = y_pred[mask]
    return np.mean(np.abs((y_true_filtered - y_pred_filtered) / y_true_filtered)) * 100
```

```python
data_spy['return'] = data_spy['Close'].pct_change() *100
data_GLD['return'] = data_GLD['Close'].pct_change() *100
data_crude_oil['return'] = data_crude_oil['Close'].pct_change() *100
data_vix['return'] = data_vix['Close'].pct_change() *100
bitcoin_data ['return']= bitcoin_data['Close'].pct_change() *100
```

```python
def condition_action_return_all_features(row):
    if row['predict'] > row['return_curr_day']:
        return 1
    elif row['predict'] < row['return_curr_day']:
        return -1
    else:
        return  0

def condition_real_action_all_features(row):
    if row['return_next_day'] > row['return_curr_day']:
        return 1
    elif row['return_next_day'] < row['return_curr_day']:
        return -1
    else:
        return  0
```
